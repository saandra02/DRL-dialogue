# RL Plots

Code for plotting average reward over time

```python
#Import libraries (required for single output and overlay plots)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Single Output Plots

```python
#Read
data = pd.read_csv("grad_1_test.txt", sep = " ", header = None) 
data.columns = ['Steps', 'Avg. Reward', 'Epsilon', 'Avg. Actions', 'Dialogues', 'Time (in Hrs)']
```

```python
print(data)
```

```python
fig, ax1 = plt.subplots() 
  
color = 'tab:blue'
ax1.set_xlabel('Learning Steps (no. experiences)') 
ax1.set_ylabel('Average Reward', color = color) 
ax1.plot(data['Steps'][:300], data['Avg. Reward'][:300], color = color)
ax1.tick_params(axis ='y', labelcolor = color) 

ax1.grid()
  
# Adding Time (in Hrs.)
ax2 = ax1.twinx() 
  
color = 'tab:green'
ax2.set_ylabel('Learning Time (in Hrs.)', color = color) 
ax2.plot(data['Steps'][:300], data['Time (in Hrs)'][:300], '--', color = color) 
ax2.tick_params(axis ='y', labelcolor = color)
  
# Adding title
plt.title('Test Period', fontweight ="bold") 
  
# Show plot
plt.show()
```

# Overlay Plots

```python
#Read data
data_sgd = pd.read_csv("sgd_test.txt", sep = " ", header = None) 
data_adadelta = pd.read_csv("adadelta_test.txt", sep = " ", header = None) 
data_adagrad = pd.read_csv("adagrad_test.txt", sep = " ", header = None) 
data_sgd.columns = ['Steps', 'Avg. Reward', 'Epsilon', 'Avg. Actions', 'Dialogues', 'Time (in Hrs)']
data_adadelta.columns = ['Steps', 'Avg. Reward', 'Epsilon', 'Avg. Actions', 'Dialogues', 'Time (in Hrs)']
data_adagrad.columns = ['Steps', 'Avg. Reward', 'Epsilon', 'Avg. Actions', 'Dialogues', 'Time (in Hrs)']
```

```python
#Plotting Avg. Rewards
fig, ax1 = plt.subplots() 
  
color = 'tab:blue'
ax1.set_xlabel('Learning Steps (no. experiences)') 
ax1.set_ylabel('Average Reward', color = color) 
ax1.plot(data_sgd['Steps'][:30], data_sgd['Avg. Reward'][:30], color = 'tab:blue', label='SGD') 
ax1.plot(data_adadelta['Steps'][:30], data_adadelta['Avg. Reward'][:30], color = 'tab:green', label='Adadelta')
ax1.plot(data_adagrad['Steps'][:30], data_adagrad['Avg. Reward'][:30], color = 'tab:red', label='Adagrad')
ax1.tick_params(axis ='y', labelcolor = color)
ax1.grid()
  
'''# Adding Time (in Hrs.)
ax2 = ax1.twinx() 
  
color = 'tab:green'
ax2.set_ylabel('Learning Time (in Hrs.)', color = color) 
ax2.plot(data['Steps'][:300], data['Time (in Hrs)'][:300], '--', color = color) 
ax2.tick_params(axis ='y', labelcolor = color) '''
  
# Adding title
plt.title('Comparing Optimizers', fontweight ="bold") 
plt.legend()
  
# Show plot
plt.show()
```
